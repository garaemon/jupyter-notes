matplotlibをjupyterで利用するには、
```
%matplotlib inline
```
を書いておく

````python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
````

sinカーブを描いてみる

````python
x = np.arange(0, 10, 0.1)
y = np.cos(x)

plt.plot(x, y)
````

gridをつけてみる

````python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.grid(True)
plt.show()
````

gridを細かくしてみる

````python
fig, ax = plt.subplots()
x_ticks = np.arange(0, 10, 1)
y_ticks = np.arange(-1.0, 1.0, 0.1)
ax.set_xticks(x_ticks)
ax.set_yticks(y_ticks)
ax.plot(x, y)
ax.grid(True)
plt.show()
````

scatterで書いてみる

````python
x = np.random.rand(100)
y = np.random.rand(100)
fig, ax = plt.subplots()
ax.scatter(x, y)
plt.show()
````

labelとtitleをつけてみる

````python
x = np.random.rand(100)
y = np.random.rand(100)
fig, ax = plt.subplots()
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_title('Foo')
ax.scatter(x, y)
plt.show()
````

凡例をつけてみる

````python
x = np.random.rand(100)
x2 = np.random.rand(100)
y = np.random.rand(100)
y2 = np.random.rand(100)
fig, ax = plt.subplots()
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_title('Foo')
ax.scatter(x, y, label='data1', c='red')
ax.scatter(x2, y2, label='data2', c='blue')
ax.legend(loc='upper left')
plt.show()
````
